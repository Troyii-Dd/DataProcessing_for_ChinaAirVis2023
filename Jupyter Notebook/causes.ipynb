```python
import pandas as pd
```

```python
meitan = pd.read_csv('./data-temp/煤炭消费量.csv', encoding='GB2312')
```

```python
jiaotan1 = pd.read_csv('./data-temp/焦炭生产量.csv', encoding='GB2312')
jiaotan2 = pd.read_csv('./data-temp/焦炭消费量.csv', encoding='GB2312')
qiyou1 = pd.read_csv('./data-temp/汽油生产量.csv', encoding='GB2312')
qiyou2 = pd.read_csv('./data-temp/汽油消费量.csv', encoding='GB2312')
```

```python
df1 = pd.melt(meitan, id_vars=['地区'], var_name='年份')
df1.columns = ['prov', 'year', 'coal']
```

```python
df2 = pd.melt(jiaotan1, id_vars=['地区'], var_name='年份')
df2.columns = ['prov', 'year', 'coke_pro']
```

```python
df3 = pd.melt(jiaotan2, id_vars=['地区'], var_name='年份')
df3.columns = ['prov', 'year', 'coke_con']
```

```python
df4 = pd.melt(qiyou1, id_vars=['地区'], var_name='年份')
df4.columns = ['prov', 'year', 'gasoline_pro']
```

```python
df5 = pd.melt(qiyou2, id_vars=['地区'], var_name='年份')
df5.columns = ['prov', 'year', 'gasoline_con']
```

```python
df = df1.merge(df2)
df = df.merge(df3)
df = df.merge(df4)
df = df.merge(df5)
```

```python
df.info()
```

```python
df['year'] = df['year'].map(lambda x: str(x)[:-1])
```

```python
df = df.sort_values(by="year", ascending=True)
```

```python
df = df.reset_index(drop=True)
```
