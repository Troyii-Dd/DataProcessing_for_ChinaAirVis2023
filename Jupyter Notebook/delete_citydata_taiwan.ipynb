```python
import pandas as pd
```

```python
df0 = pd.read_csv('./hourly.csv', encoding='utf-8')
df1 = pd.read_csv('./daily.csv', encoding='utf-8')
df2 = pd.read_csv('./monthly.csv', encoding='utf-8')
df3 = pd.read_csv('./yearly.csv', encoding='utf-8')
```

```python
df1.loc[df1['province'].str.contains('台湾')]
```

```python
df0 = df0.loc[~df0['province'].str.contains('台湾')]
df1 = df1.loc[~df1['province'].str.contains('台湾')]
df2 = df2.loc[~df2['province'].str.contains('台湾')]
df3 = df3.loc[~df3['province'].str.contains('台湾')]
```

```python
df0.to_csv('hourly_new.csv', encoding='utf-8-sig', index = False)
df1.to_csv('daily_new.csv', encoding='utf-8-sig', index = False)
df2.to_csv('monthly_new.csv', encoding='utf-8-sig', index = False)
df3.to_csv('yearly_new.csv', encoding='utf-8-sig', index = False)
```
